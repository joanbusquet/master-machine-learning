<h1>1. Cómo crear una array unidimensional</h1>

```python
import numpy as np

# Creamos un np array de rango 1 (unidimensional, una sola fila)
unidimensional_array = np.array([7,77,777, 1234]) # Le pasamos los valores a la función de np.array
print(type(unidimensional_array)) # ¿De que tipo de clase es?
```

<h1>4. Array de índices</h1>

```python
# Creamos un nuevo array multidimensional
array_a = np.array([[11,12,13], [21,22,23], [31,32,33], [41,42,43]])
print(array_a)
print(array_a.shape)
```

```python
# Creamos 2 arrays con números enteros que usaremos como índices
cols = np.array([0,1,2,0])
rows = np.arange(4) # genera secuencia numeros consecutivos 0 1 2 3

print("Elegimos estos índices para filas: ", rows)
print("Elegimos estos índices para columnas: ", cols)
```

<h1>5. Indexado Booleano</h1>
<h3>Indexado booleano para cambiar elementos</h3>

```python
# Creamos array de 3 filas x 2 columnas
array_b = np.array([[11,12], [21,22],[31,32]])
print(array_b)
print(array_b.shape)
```

```python
# 1 PASO: crear un filtro de búsqueda en numpy
filtro = (array_b > 15) # Va a tomar valor True en todos los valores que cumplan la condicion
print(filtro) # Mostramos el filtro
```

```python
# 2 PASO: filtramos el array obteniendo solo los valores True del filtro anterior
filtrado = array_b[filtro] 
print(filtrado)
```

```python
# SOLO 1 PASO: No es necesario primero crear el filtro y despues aplicarlo, podemos hacerlo al unísono
print(array_b[(array_b > 15)])

# Otro filtro
print(array_b[(array_b % 2 == 0)]) # Si es multiplo de 2
```

<h3>El uso principal de este tipo de filtros es de poder modificar los valores del array que cumplan cierto criterio</h3>

```python
# Vamos a obtener todos los valores multiplos de 2 y les sumaremos 1000
array_b[(array_b % 2 == 0)] += 1000

# Mostramos resultado con la modificación aplicada
print(array_b)
```

<h1>6. Recortando arrays (Slice)</h1>
<h3>Hacer Slice a un array sirve para aislar una "sub-región" de la matriz de un ndarray. Una vez hecho un slice, se genera un array recortado (un puntero) manteniendo los mismos valores.</h3>

```python
# Creamos array multidimensional de 4 filas por 6 columnas
un_array = np.array([[11,12,13,14,15,16], [21,22,23,24,25,26], [31,32,33,34,35,36], [41,42,43,44,45,46]])
print(un_array)
```

```python
# Creamos un slice de 2x2 del array "un_array"
# :2 es igual a 0:2
# :2 es del eje de filas (el ultimo digito (fila 2, no está incluido!)
# 1:3 es del eje de columnas (el ultimo digito (columna 3, no está incluido!)
un_slice = un_array[:2, 1:3] 
print(un_slice)
```

<h3>Resultado de un_array[:2, 1:3]</h3>



```python
# Imprimimos el valor fila 0 columna 0 del slice
# Como vemos en la salida, el slice genera un nuevo índice, nace un array nuevo
print("Slice: ", un_slice[0,0])
```

```python
# Sumaremos 100 al valor de la posición 0,0 del slice
un_slice[0,0] += 100
print("Slice un_slice:", un_slice[0, 0])
```

```python
# Imprimimos la posicion equivalente de la posicion modificada del slice pero en la matriz original
print("Array original un_array:",un_array[0,1])
```

<h3>IMPORTANTE: Si se modifica un valor en alguna posición del slice, también se cambia a la matriz original! Es un puntero!</h3>

<h3>Crear un nuevo array a partir de un slice (PUNTERO NO!)</h3>

```python
# Creamos un nuevo array a partir del contenido del slice que hemos hecho previamente
nuevo_array = np.array(un_slice)
print(nuevo_array)
```

```python
# Seleccionar una sola fila, slice unidimensional
row_rank1 = un_array[1, :]
print(row_rank1, row_rank1.shape)
```

```python
# Seleccionar una sola columna, slice unidimensional
row_rank2 = un_array[:, 1]
print(row_rank2, row_rank2.shape)
```

<h1>8. Operaciones Aritmeticas</h1>

```python
# Creamos dos arrays para hacer operaciones
x = np.array([[111,112], [121,122]], dtype=np.int)
y = np.array([[211.1,212.1], [221.1,222.1]], dtype=np.float64)

print(x)
print()
print(y)
```

```python
# RESTA (todos los valores de dentro de la matriz)
print(x - y) 
print()
print(np.subtract(x,y)) # Método de numpy que resta valores de matrices
```

```python
# MULTIPLICACION (todos los valores de dentro de la matriz)
print(x * y)
print()
print(np.multiply(x,y)) # Método de numpy que multiplica valores de matrices
```

```python
# DIVISION (todos los valores de dentro de la matriz)
print(x / y)
print()
print(np.divide(x, y)) # Método de numpy que multiplica valores de matrices
```

```python
# RAÍZ CUADRADA
print(np.sqrt(x))
```

<h1>9. Operaciones Estadísticas Básicas</h1>

```python
# Generar array aleatorio de 2 filas y 5 columnas
# Los valores 2 y 5 pasados como argumentos en el método anterior son los valores para fila y columna de la matriz
# El método random.randn devuelve valores aleatorios que pertenecen a una distribución normal N(0,1). 
# Distribución aleatoria centrada en 0 y con desvío estándar 1.
arr = 10 * np.random.randn(2,5) # Multiplicamos 10
print(arr)
```

```python
# Calcular la media (promedio de un array)
print(arr.mean()) # Función mean de python (sumar todos los valores y los divide por 10)
```

```python
# Calcular el promedio, pero en este caso, fila por fila
# IMPORTRANTE:
# Eje 0 = columnas
# Eje 1 = filas
print(arr.mean(axis = 1)) # Le pasamos el parametro axis = 1 (eje 1, fila) | Devuelve 2 promedio (de las dos filas del array)
```

```python
# Calcular el promedio, pero en este caso, columna por columna
# IMPORTRANTE:
# Eje 0 = columnas
# Eje 1 = filas
print(arr.mean(axis = 0)) # Le pasamos el parametro axis = 0 (eje 0, columna) | Devuelve 5 promedios (de las dos filas del array)
```

```python
# Creamos array de 10 elementos aleatorios ()
desordenado = np.random.randn(10) # Le pasamos como parametro el numero de elementos que queremos generar dentro de la array
print(desordenado)
```

```python
# Creamos una copia de la array
ordenado = np.array(desordenado)

# Lo ordenamos
ordenado.sort() # Usamos el método sort de python para ordenar de menor a mayor
print(ordenado)
```

```python
# Buscando elementos únicos (MUY ÚTIL!!!)
array = np.array([1,2,1,4,2,1,4,2])
print(np.unique(array))
```

```python
# Operaciones de conjunto
s1 = np.array(['escritorio', 'silla', 'mesa'])
s2 = np.array(['escritorio', 'silla', 'lampara'])
print(s1)
print(s2)
```

```python
# Intersección (¿Qué valores tienen en común?) - 1d = 1 dimensión de array
print(np.intersect1d(s1,s2))
```

```python
# Unión de arrays 1d = unidimensionales (NOTA: si hay algun duplicado, solo se mantiene 1)
print(np.union1d(s1,s2))
```

```python
# Elementos de s1 que NO EXISTEN en s2
print(np.setdiff1d(s1,s2))
```

```python
# ¿Que elementos de s1 EXISTEN en s2? Devuelve Bool
print(np.in1d(s1,s2))
```

<h3>Uniendo data sets (arrays multidimensionales)</h3>

```python
# Creamos 2 array randoms
# 1 parametro (low): Número mínimo a generar
# 2 parametro (high): Número máximo a generar
# 3 parametro (size): Tamaño del array
k = np.random.randint(low=2,high=50,size=(2,2)) 
print(k)
print()

m = np.random.randint(low=2,high=50,size=(2,2))
print(m)
```

```python
# Unimos "aplilando verticalmente": mismas columnas, más filas
print(np.vstack((k,m))) # Vertical stack
```

```python
# Unimos "aplilando horizontalmente": más columnas, mismas filas
print(np.hstack((k,m))) # Horizontal stack
```

```python
# Unir contatenando (lo mismo que el anterior pero indicando el eje (axis))
# Eje 0 = columnas
# Eje 1 = filas
print(np.concatenate([k,m], axis = 0)) # Columnas
print()

print(np.concatenate([k,m], axis = 1)) # Filas
```

<h1>11. Propagación</h1>
<h3>Para más detalles: <a href="https://docs.scipy.org/doc/numpy-1.10.1/user/basics.broadcasting.html">https://docs.scipy.org/doc/numpy-1.10.1/user/basics.broadcasting.html</a></h3>


```python
# Creamos array de 4x4 con 0
base = np.zeros((4,4))
print(base)
```

```python
# Creamos array unidimensional
row = np.array([1, 0, 2, 7])
print(row)
print(row.shape)
```

```python
# PROPAGACIÓN VERTICAL
# Sumamos la array unidimensional a la multidimensional de 4x4 para ver que se propaga (tiene que tener el mismo ancho!)
# Se propaga verticalmente fila por fila
y = base + row
print(y)
```

```python
# Creamos array de 1 fila
col = np.array([[1, 0, 2, 7]]) #se pone corchete adicional para poder aplicar la función T y convertirlo en una columna
print(col)
print(col.shape)
```

```python
# Verticalizamos la fila
col = col.T # T es una función para verticalizar una fila
print(col)
```

```python
# PROPAGACIÓN HORIZONTAL
# Sumamos la array unidimensional a la multidimensional de 4x4 para ver que se propaga (tiene que tener el mismo ancho!)
# Se propaga horizontalmente columna por columna base + row
y = base + col
print(y)
```

```python
#Prueba de propagación de una array más pequeña que la multidimensional
col = np.array([[0,1,2]]) #creamos array de 3 posiciones
col = col.T # verticalizamos array
print(col)
print()

# Veamos que pasa cuando hacemos una propagación de arrays de distintos tamaños:
y = base + col
print(y)

# ERROR! ValueError: operands could not be broadcast together with shapes (4,4) (3,1) 
```

```python
# Prueba de propagación de una array de 1 solo elemento
arr_x = np.array([1])
print(base + arr_x)

# FUNCIONA! Se propagan todos los elementos con el valor de la array de 1 solo elemento
```

<h1>EXTRA: Prueba de rendimiento (nd_array vs lista normal) </h1>

```python
from numpy import arange
from timeit import Timer

size    = 1000000
timeits =  1000

# Creamos nd_array de 1 millon de registros
nd_array = arange(size) 
print("Tipo de dato: ",type(nd_array))
print("Forma: ", nd_array.shape)
```

```python
# Con la funcion timer podemos medir el tiempo que lleva realizar una operación o función
timer_numpy = Timer("nd_array.sum()", "from __main__ import nd_array")
print("Tiempo que tarda numpy array: %f segundos" % (timer_numpy.timeit(timeits)))
```

```python
# Creamos lista de Python de 1 millon de registros
a_list = list(range(size))
print("Tipo de dato: ",type(a_list))
print("Forma/tamaño: ",len(a_list))
```

```python
timer_list = Timer("sum(a_list)", "from __main__ import a_list")
print("Tiempo que tarda lista de Python: %f segundos" % (timer_list.timeit(timeits)))
```
