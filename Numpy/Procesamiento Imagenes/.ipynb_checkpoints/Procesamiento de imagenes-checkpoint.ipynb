<h1>4. Te칩rica: Creamos una m치scara circular</h1>
<h3>Crear mascara circular y poner dentro la imagen</h3>
<h3>https://docs.scipy.org/doc/numpy/reference/generated/numpy.ogrid.html

<p>Para obtener el punto central de la matriz, dividimos las las filas entre 2 y columnas entre 2</p>
<p>Por ejemplo, una matriz de 10 x 16, el punto central ser칤a 5 x 8</p>

<p>Para conseguir hacer la redonda entorno al punto centrar realizaremos los siguientes calculos:</p>
<ul>
    <li>1. El primer paso es calcular la distancia entre cada casilla y la casilla central</li>
    <li>1.1. Generamos una matriz calculando la distancia de todas las filas (EJE X) hacia la fila central</li>
</ul>
<p><i>Formula filas: Distancia = (X - Centro de X)</i>
<p>Vemos que existen numeros negativos, para solventar esto lo elevamos al cuadrado. </p>
<p><strong>Formula filas: Distancia = (X - Centro de X) ** 2</strong></p>
<br />
<p>La idea es obtener los valores en referencia al centro, lo podriamos elevar al cubo a lo que se quiera</p>

<ul>
    <li>1.2. Generamos una matriz calculando la distancia de todas las columnas (EJE Y) hacia la fila central</li>
</ul>
<p><i>Formula filas: Distancia = (Y - Centro de Y)</i>
<p>Vemos que existen numeros negativos, para solventar esto lo elevamos al cuadrado. </p>
<p><strong>Formula columnas: Distancia = (Y - Centro de Y) ** 2</strong>
<p>La idea es obtener los valores en referencia al centro, lo podriamos elevar al cubo a lo que se quiera</p>
<ul>
    <li>2. Sumamos los valores de ambas tablas y nos da el resultado final</li>
</ul>


<h1>4. Pr치ctica: Creamos una m치scara circular en Python</h1>

```python
# Guardamos las medidas de la imagen que tenemos cargada previamente en variables
total_rows, total_cols, total_layers = image1.shape # Shape retorna (900, 1600, 3 y se guardan en las variables con el mismo orden)

# La funci칩n ogrid genera dos valores, X es una matriz en COLUMNA unidimensional de 0 a 899
# e Y es una matriz FILA unidimensional de 0 a 1599
# Es ideal para crear los 칤ndices de los ejes
X, Y = np.ogrid[:total_rows, :total_cols]

print("X: ",X.shape, " Y: ",Y.shape)
```

```python
# Calulamos la fila central (dividimos las filas entre 2 y las columnas entre 2)
center_row, center_col = total_rows / 2, total_cols / 2

print("Fila central: ",center_row, "Columna central: ",center_col)
```

```python
# Creamos una matriz con las formulas calculadas en la teoria y las ponemos en la nueva matriz
# Distancia del centro = (X - center_row)**2 + (Y - center_row)**2
dist_from_center = (X - center_row)**2 + (Y - center_col)**2

print("Matriz que contiene las distancias del centro: ")
print(dist_from_center)

print("Distancia maxima de los extremos de la imagen hasta el centro es: ",np.amax(dist_from_center))
print("Distancia minima de los extremos de la imagen hasta el centro es: ",np.amin(dist_from_center))
```

```python
# Calculamos el tama침o del radio deseado
# Queremos que se mantenga la altura pero que los laterales se corten, para ello dividimos el alto entre 2
# IMPORTANTE: Lo elevamos al cuadrado ya que para convertir las distancias del centro en el paso anterior tambi칠n lo hemos hecho!
radius = (total_rows / 2)**2 # 900 / 2 = 450  | 450**2 = 202500
print("El radio es: ",radius)
```

```python
# Creamos un filtro para la mascara circular con la matriz de distancias y el radio
circular_mask = (dist_from_center > radius) # Todo lo que est칠 fuera del circulo valdr치 True y lo que est칠 dentro ser치 False

# Imprimimos el tama침o del filtro (el tama침o del shape debe ser igual que el de la imagen)
print(circular_mask.shape)

# Imprimimos una parte de la imagen que es el borde para ver el cambio
print(circular_mask[300:600, 600:1000]) 

# Imprimimos la matriz del filtro
print(circular_mask)
```

```python
# Aplicamos la m치scara en la foto (a partir de la matriz del filtro creado en el paso anterior)
image1[circular_mask] = 0 # Convertimos todos los True del filtro (lo que est치 fuera del c칤rculo) a color negro (0)

# Mostramos la imagen por pantalla con la libreria Pyplot
plt.figure(figsize=(10,20))
plt.imshow(image1)
```

<h1>5. Filtro de color</h1>

```python
# Cargamos una imagen de Internet
url = "https://3.bp.blogspot.com/-4lNOY6d8Q3E/Vc5lNi73qMI/AAAAAAABxdQ/7RN64mKqrxs/s1600/Pedraforca254.jpg"
internet_image = Image.open(urllib.request.urlopen(url))

# Creamos matriz a partir de la imagen
image = np.array(internet_image)

# Mostramos la imagen por pantalla con la libreria Pyplot
plt.figure(figsize=(10,20))
plt.imshow(internet_image)
```

```python
# Creamos un filtro que tomar치 de valor True d칩nde existan valores mayores a 115, "de 115-255 es verde intenso" (En la dimensi칩n G (verde))
green_mask = image[:,:,1] > 115 # Alto : (todo) | Ancho : (todo) | Canal: 1 (G verde)

# Mostramos la matriz booleana del filtro
print("Filtro de la mascara verde: ")
print(green_mask)
```

```python
# Operaremos sobre los p칤xeles a donde hay verdes intensos (en las 3 dimensiones RGB si se cumple el filtro)
image[green_mask, 0] = 0 # Todos se cumpla el filtro de la dimensi칩n R se pondr치 a 0
image[green_mask, 1] = 0 # Todos se cumpla el filtro de la dimensi칩n G se pondr치 a 0
image[green_mask, 2] = 0 # Todos se cumpla el filtro de la dimensi칩n B se pondr치 a 0

print(image[green_mask])
```

```python
# Mostramos la imagen por pantalla con la libreria Pyplot
plt.figure(figsize=(10,20))
plt.imshow(image)
```
